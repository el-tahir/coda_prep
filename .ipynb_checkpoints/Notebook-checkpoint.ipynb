```python
import csv
import pandas as pd
import json

def connect(sent):
    chars = ['ح', 'ب','ف']
    
    split = sent.split()
    
    for i in range(len(split)):
        try:
            if split[i] in chars:
                split[i+1] = f'{split[i]}{split[i+1]}'
                del split[i]
        except IndexError:
            break
    
    return ' '.join(x for x in split)

word_gloss = {
        'ل':'لي',
        'ول':'ولا',
        'الى':'لي',
        'م':'ما',
        'دا':'ده',
        'كدا':'كده',
        'ي':'يا'
        }

char_gloss = {
    'ئ':'ي'
}
def word_level(sent):
    new = []
    for word in sent.split():
        if word in word_gloss:
            new.append(word_gloss[word])
        else:
            new.append(word)
           
    return ' '.join(new)

def char_level(word):
    for c in char_gloss:
        if c in word:
            i = word.index(c)
            return word[:i] + char_gloss[c] + word[i+len(char_gloss[c]):]
    if word[-1] == 'ء':
        return word[:-1]
    return word

def codafy(sent):
    ans = word_level(sent)
    ans = ' '.join(char_level(word) for word in ans.split())
    return connect(ans)



            


```

```python
frame = pd.read_csv('1600data.csv')

frame.drop('Unnamed: 0', axis = 1, inplace = True)

df_short = pd.DataFrame(columns= ['sent', 'CODA','eltahir'])
df_long = pd.DataFrame(columns = ['sent', 'CODA', 'eltahir'])
df_same = pd.DataFrame(columns = ['sent', 'CODA','eltahir'])

total_same = 0
total_short = 0
total_long = 0

coda_good_same = 0


eltahir_good_same = 0
eltahir_good_short = 0
eltahir_good_long = 0

for i in range(len(frame)):
    raw = frame['sent'].values[i]
    coda = frame['CODA'].values[i]
    
    
    if (len(raw.split()) == len(coda.split())):
        df_same.loc[len(df_same)] = [raw, coda, codafy(raw)]
        
        if raw != coda:
            total_same += 1
            if coda == codafy(raw):
                eltahir_good_same += 1
       
    if (len(raw.split()) > len(coda.split())):
        df_short.loc[len(df_short)] = [raw, coda, codafy(raw)]
        if raw != coda:
            total_short += 1
            if coda == codafy(raw):
                eltahir_good_short += 1
        
    if (len(raw.split()) < len(coda.split())):
        df_long.loc[len(df_long)] = [raw, coda, codafy(raw)]
        if raw != coda:
            total_long += 1
        if coda == codafy(raw):
            eltahir_good_long += 1
        
same_change = pd.DataFrame(columns = ['sent', 'CODA', 'eltahir'])


print("Sentences which:\n")
print("stayed same length:")
print(f"Total changes: {total_same}, eltahir did {eltahir_good_same}, accuracy : {100*eltahir_good_same/total_same:.1f}\n")
print("made shorter")
print(f"Total changes: {total_short}, eltahir did {eltahir_good_short}, accuracy : {100*eltahir_good_short/total_short}\n")
print("made longer")
print(f"Total changes: {total_long}, eltahir did {eltahir_good_long}, accuracy : {eltahir_good_long*100/total_long}\n")

print(f"total sentences that need change = {total_long+total_short+total_same}")
```
