```python
import csv
import pandas as pd

df = pd.read_csv('data.csv')


for i in range(1600, 2000):
    df.drop(i, inplace = True)
    
```

```python
df.drop(['Unnamed: 0', 'split', 'lang'],axis = 1, inplace = True)
```

```python
for i in range(len(df)):
    if not pd.isnull(df['revised'].values[i]):
        df['sent'].values[i] = df['revised'].values[i]
df.drop(['revised', 'sentID.BTEC'], axis = 1, inplace = True)

hamzas = ['أ','إ']
punc = ['؟','،','.']

for i in range(len(df)):
    
    df['CODA'].values[i] = df['CODA'].values[i].translate({ord(i): None for i in punc})
    
    raw = df['sent'].values[i]
    new = []
    
    for word in raw.split():
        
        if word[0] in hamzas:
            new_word = 'ا'
            for l in range(1, len(word)):
                new_word += word[l]
            new.append(new_word)
        else:
            new.append(word)    
        
    df['sent'].values[i] = (' '.join(new)).translate({ord(i): None for i in punc})
            

    
```

```python
df.to_csv('1600data.csv')
```
